```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Users and products
users = [f"u{i}" for i in range(1, 101)]  #100 fake users create
products = {
    "watch_001": "watches",
    "jacket_002": "clothing",
    "shoe_003": "footwear",
    "phone_004": "electronics",
    "book_005": "books"
}

#behavior data
events = ["view", "add_to_cart", "purchase"]

data = []

start_time = datetime.now()

for _ in range(2000):  # more data = better training
    user = random.choice(users)
    product = random.choice(list(products.keys()))
    category = products[product]
    price = random.randint(300, 10000)

    # realistic behavior (IMPORTANT)
    event = random.choices(
        events,
        weights=[0.65, 0.25, 0.10]  # mostly views, fewer purchases
    )[0]

    time = start_time + timedelta(minutes=random.randint(0, 10000))
    session = f"s{random.randint(1,500)}"

    data.append([user, product, category, price, event, time, session]) #complete record

# Create DataFrame(list--> table)
df = pd.DataFrame(data, columns=[
    "user_id", "product_id", "category", "price",
    "event_type", "timestamp", "session_id"
])

# Save dataset
df.to_csv("smartbazaar_data.csv", index=False)

print("Dataset created successfully ✅")
df.head()
```

# Data Cleaning

```python
df.dropna(inplace=True) #remove missing values

df['timestamp'] = pd.to_datetime(df['timestamp']) #datetime format for timestamp
```

#ML

```python
#Time-based feature

df['hour'] = df['timestamp'].dt.hour
```

```python
#Target variable

df['label'] = df['event_type'].apply(lambda x: 1 if x == 'purchase' else 0)

```

```python
#Encode category

df = pd.get_dummies(df, columns=['category'])
```

```python
df = df.astype({col: 'int' for col in df.columns if 'category_' in col})
```

```python
df.head()
```

```python
df['label'].value_counts()
```

#TRAIN MODEL

```python
# Step 1: Prepare X and y

X = df.drop(['event_type', 'label', 'timestamp', 'user_id', 'session_id'], axis=1)
y = df['label']
```

```python
#Step 2: Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#train
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    class_weight={0:1, 1:5},  # boost buyers importance
    n_estimators=200,
    random_state=42
)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
